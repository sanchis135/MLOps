```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Load the cleaned dataset

```python
data_path = "../data/processed/flights_clean.csv"
df = pd.read_csv(data_path)
```

# Basic information

```python
print("üßæ Dataset Shape:", df.shape)
print("\nüìå Columns:\n", df.columns.tolist())
print("\n‚ùì Missing Values:\n", df.isnull().sum())
print("\nüìä Data Types:\n", df.dtypes)
print("\nüéØ Class Balance (is_delay):\n", df["is_delay"].value_counts(normalize=True))
```

# Summary statistics

```python
display(df.describe())
```

# Set Seaborn style

```python
sns.set(style="whitegrid")
```

# Arrival Delay Distribution

```python
plt.figure(figsize=(12, 6))
plt.subplot(1, 2, 1)
sns.histplot(df['ArrDelay'], bins=50, kde=True)
plt.title('Arrival Delay Distribution')
plt.xlabel('Arrival Delay (minutes)')
```

# Flight Distance Distribution

```python
plt.subplot(1, 2, 2)
sns.histplot(df['Distance'], bins=50, kde=True, color="orange")
plt.title('Flight Distance Distribution')
plt.xlabel('Distance (miles)')

plt.tight_layout()
plt.show()
```

# Categorical Feature Analysis

```python
# List of categorical features to analyze
categorical_features = [
    "Reporting_Airline", "Origin", "OriginState", "Dest", "DestState", 
    "DayOfWeek", "Month", "CRSDepHour"
]

# Plot delay rate for each categorical feature
for col in categorical_features:
    plt.figure(figsize=(10, 5))
    order = df.groupby(col)["is_delay"].mean().sort_values(ascending=False).index
    sns.barplot(data=df, x=col, y="is_delay", order=order, palette="coolwarm")
    plt.title(f"Delay Rate by {col}")
    plt.ylabel("Proportion of Delays")
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

# Correlation Analysis (Numerical Features)

```python
# Select numerical features
numerical_features = ["Distance", "ArrDelay", "ArrDelayMinutes", "AirTime", "CRSDepHour"]

# Correlation matrix
plt.figure(figsize=(8, 6))
corr = df[numerical_features + ["is_delay"]].corr()
sns.heatmap(corr, annot=True, cmap="coolwarm", fmt=".2f", square=True)
plt.title("Correlation Matrix")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,8))
num_cols = ['Year', 'Quarter', 'Month', 'DayofMonth', 'DayOfWeek', 'CRSDepHour', 'Cancelled', 'Diverted',
            'Distance', 'ArrDelay', 'ArrDelayMinutes', 'AirTime', 'is_delay']
corr = df[num_cols].corr()
sns.heatmap(corr, annot=True, fmt=".2f", cmap='coolwarm', center=0)
plt.title('Correlation Heatmap of Numeric Features')
plt.show()
```

# Seasonality Analysis - Delay Rate by Month

```python
plt.figure(figsize=(10, 5))
month_order = sorted(df["Month"].unique())
sns.barplot(data=df, x="Month", y="is_delay", order=month_order, palette="viridis")
plt.title("Delay Rate by Month")
plt.ylabel("Proportion of Delays")
plt.xlabel("Month")
plt.xticks(ticks=range(0, 12), labels=[
    "Jan", "Feb", "Mar", "Apr", "May", "Jun",
    "Jul", "Aug", "Sep", "Oct", "Nov", "Dec"
])
plt.tight_layout()
plt.show()
```

# Seasonality Analysis - Delay Rate by Day of Week

```python
plt.figure(figsize=(10, 5))
dow_order = [1, 2, 3, 4, 5, 6, 7]  # Monday to Sunday
sns.barplot(data=df, x="DayOfWeek", y="is_delay", order=dow_order, palette="magma")
plt.title("Delay Rate by Day of Week")
plt.ylabel("Proportion of Delays")
plt.xlabel("Day of Week")
plt.xticks(ticks=dow_order, labels=[
    "Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"
])
plt.tight_layout()
plt.show()
```

# Histogram / bar plot of is_delay distribution

```python
plt.figure(figsize=(6,4))
sns.countplot(x='is_delay', data=df)
plt.title('Distribution of Flight Delays (is_delay)')
plt.xlabel('Is Delayed (0 = No, 1 = Yes)')
plt.ylabel('Count')
plt.show()
```

# Delay rate by airline (top 10 airlines by flight count)

```python
plt.figure(figsize=(12,5))
top_airlines = df['Reporting_Airline'].value_counts().nlargest(10).index
delay_rate_airline = df[df['Reporting_Airline'].isin(top_airlines)].groupby('Reporting_Airline')['is_delay'].mean()
sns.barplot(x=delay_rate_airline.index, y=delay_rate_airline.values)
plt.title('Average Delay Rate by Top 10 Airlines')
plt.xlabel('Airline')
plt.ylabel('Average Delay Rate')
plt.show()
```

# Delay rate by origin airport (top 10 airports by flight count)

```python
plt.figure(figsize=(12,5))
top_origins = df['Origin'].value_counts().nlargest(10).index
delay_rate_origin = df[df['Origin'].isin(top_origins)].groupby('Origin')['is_delay'].mean()
sns.barplot(x=delay_rate_origin.index, y=delay_rate_origin.values)
plt.title('Average Delay Rate by Top 10 Origin Airports')
plt.xlabel('Origin Airport')
plt.ylabel('Average Delay Rate')
plt.show()
```
